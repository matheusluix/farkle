### PARTE II: Implementação do Jogo _Farkle_ em Python Funcional

__Farkle__ é um jogo envolvendo seis dados, conhecido desde o século XVII. Sua origem não é bem estabelecida, embora se saiba que ele era bem comum nas grandes navegações. Enquanto há um conjunto bem estabelecido de regras gerais, há uma grande variedade de formas de pontuação. Nesta seção você vai implementar um programa que simula uma partida de Farkle entre dois jogadores, usando Python funcional. A escrita do programa vai se dar através da escrita de várias funções intermediárias. Utilize _apenas_ programação funcional, conforme visto em sala:

* Sem laços for ou while imperativos (use recursão, list comprehensions, map, filter, reduce).
* Sem mutação de variáveis (crie novas listas em vez de modificar as existentes).
* Funções devem ser puras sempre que possível.
* Sua solução deve ser recursiva sempre que isso for solicitado explicitamente. 

Em seguida, temos a descrição do jogo a ser implementada.

* O jogador começa com 6 dados.
* Ele lança todos os dados disponíveis (na primeira jogada, os 6).
* Após o lançamento, ele deve separar pelo menos um dado que pontue (de acordo com a tabela de pontuação). Os dados separados são "guardados" e somam pontos na rodada.
* Os dados não separados (que não pontuaram ou que ele optou por não guardar) podem ser relançados se o jogador quiser continuar.
* Se o jogador conseguir usar todos os 6 dados em combinações pontuadoras (ou seja, separar todos), ele pode continuar rolando todos os 6 novamente (como se tivesse uma nova mão), acumulando pontos sucessivamente.
* Se em qualquer lançamento nenhum dado pontuar (Farkle), o jogador perde todos os pontos acumulados na rodada e passa a vez.

Pontuação:
* Cada 1 vale 100 pontos.
* Cada 5 vale 50 pontos.
* Trinca de 1: 1000 pontos.
* Trinca de 2 a 6: 100 × o número (ex: trinca de 2 = 200, trinca de 3 = 300, etc.).
* Sequência de 1 a 6: 1500 pontos.
* Se nenhum dado pontuar na rolagem (Farkle), o jogador perde todos os pontos acumulados na rodada.

__Objetivo do Jogo__: ser o primeiro a alcançar 10 ou mais pontos ao fim de uma rodada completas.

#### 2.2. Contar ocorrências
Crie a função ```contar(lst, alvo)``` que retorna o número de vezes que _alvo_ ocorre em _lst_, usando recursão.
Exemplo:
```python
contar([1, 5, 1, 1, 5, 6], 1) -> 3
contar([1, 5, 1, 1, 5, 2], 2) -> 1
```

````python
def contar(lst, alvo):
    if len(lst) == 0:
        return 0
    
    total = 0
    for i in range(len(lst)):
        if lst[i] == alvo:
            total += 1
    
    return total
print(contar([1, 5, 1, 1, 5, 6], 1))
print(contar([1, 5, 1, 1, 5, 2], 2))
````

#### 2.4. Pontos por número e dados pontuadores
Crie a função ```pontos_por_numero_(dados, num)``` que considerando um arranjo dos seis dados, retorna a pontuação referente a _num_ e quantos foram os dados pontuadores. Considere que o arranjo nunca pode ser formado pelos números de 1 a 6, cada um deles ocorrendo uma única vez, como em [1, 2, 3, 4, 5, 6]. 
Exemplo:
```python
pontos_por_numero([1, 2, 3, 4, 5, 6], 1) -> (100, 1) # O numero 1 ocorre 1x no arranjo. Portanto, para 1, a pontuação do arranjo é 100 com 1 dado pontuando.
pontos_por_numero([1, 2, 3, 4, 5, 6], 5) -> (50, 1) # O numero 5 ocorre 1x no arranjo. Portanto, para 5, a pontuação do arranjo é 50 com 1 dado pontuando.
pontos_por_numero([1, 5, 1, 1, 5, 6], 1) -> (1000, 3) # O numero 1, ocorre 3x no arranjo. Portanto, para 1, a pontuação do arranjo é 1000 com 3 dados pontuando.
pontos_por_numero([1, 5, 1, 1, 5, 6], 5) -> (100, 2) # O numero 5, ocorre 2x no arranjo. Portanto, para 5, a pontuação do arranjo é 100 com 2 dados pontuando.
pontos_por_numero([1, 5, 1, 1, 5, 6], 6) -> (0, 0) # O numero  6, ocorre 1x no arranjo. Portanto, para 6, a pontuação do arranjo é 0 com nenhum dado pontuando.
pontos_por_numero([1, 1, 1, 1, 1, 1], 1) -> (2000, 6) # O numero 1, ocorre 6x no arranjo. Portanto, para 1, a pontuação do arranjo é 2000 com 6 dados pontuando.
pontos_por_numero([5, 5, 5, 5, 5, 5], 5) -> (1000, 6) # O numero 5, ocorre 6x no arranjo. Portanto, para 5, a pontuação do arranjo é 1000 com 6 dados pontuando.
pontos_por_numero([2, 2, 2, 2, 2, 2], 2) -> (400, 6) # O numero 2, ocorre 6x no arranjo. Portanto, para 2, a pontuação do arranjo é 400 com 6 dados pontuando.
```

````python
def pontos_por_numero(dados, num):
    contagem = contar(dados, num)
    
    def calcular_trincas(n):
        return n // 3, n % 3

    trincas, resto = calcular_trincas(contagem)
    
    return (
        (trincas * 1000 + resto * 100, contagem) if num == 1 else
        (trincas * 500 + resto * 50, contagem) if num == 5 else
        (trincas * (num * 100), trincas * 3)
    )

print(pontos_por_numero([1, 2, 3, 4, 5, 6], 1))
print(pontos_por_numero([1, 2, 3, 4, 5, 6], 5))
print(pontos_por_numero([1, 5, 1, 1, 5, 6], 1))
print(pontos_por_numero([1, 5, 1, 1, 5, 6], 5))
print(pontos_por_numero([1, 5, 1, 1, 5, 6], 6))
print(pontos_por_numero([1, 1, 1, 1, 1, 1], 1))
print(pontos_por_numero([5, 5, 5, 5, 5, 5], 5))
print(pontos_por_numero([2, 2, 2, 2, 2, 2], 2))
````

#### 2.5. Pontuação da jogada
Crie a função ```pontuacao_jogada_(dados)``` que considerando um arranjo dos seis dados, retorna a pontuação referente ao arranjo bem como o número de dados pontuadores. Defina ```pontuacao_jogada``` em função de ```pontos_por_numero```. 
Exemplo:
```python
pontuacao_jogada([1, 5, 2, 3, 4, 6]) -> (1500, 6) # todos os números de 1 a 6, pontuação de 1500 com todos os dados (6) usados em pontuação
pontuacao_jogada([2, 2, 3, 4, 4, 6]) -> (0, 0) # 0 pontos, nenhum dado usado em pontuação
pontuacao_jogada([1, 1, 1, 2, 3, 4]) -> (1000, 3) # 3 dados pontuaram (a trinca de 1 que resulta em 1000 pontos) 
pontuacao_jogada([2, 2, 2, 4, 5, 6]) -> (250, 4) # 4 dados pontuaram (a trinca de 2 e o numero 5, resultando em 250 pontos) 
pontuacao_jogada([1, 5, 2, 2, 3, 4]) -> (150, 2) # 2 dados pontuaram (1 e 5, resultando em 150 pontos) 
pontuacao_jogada([2, 2, 2, 2, 2, 2]) -> (400, 6) # 6 dados pontuaram (2 trincas de 2, resultando em 400 pontos)
pontuacao_jogada([2, 2, 2, 3, 3, 3]) -> (500, 6) # 6 dados pontuaram (trinca de 2 e trinca de 3, resultando em 500 pontos)
pontuacao_jogada([4, 4, 4, 6, 6, 6]) -> (1000, 6) # 6 dados pontuaram (trinca de 4 e trinca de 6, resultando em 1000 pontos)
pontuacao_jogada([5, 6, 5, 6, 6, 6]) -> (700, 5) # 5 dados pontuaram (dois 5 e trinca de 6, resultando em 700 pontos) -- note que um dos 6 não pontuou
```

````python
def pontuacao_jogada(dados):
    if sorted(dados) == [1,2,3,4,5,6]:
        return (1500,6)

    def extrai_faces_roladas(n):
        return valores_unicos_(n)

    faces_presentes = extrai_faces_roladas(dados)

    return reduce(
        lambda acumulador, atual: (acumulador[0] + atual[0], acumulador[1] + atual[1]),
        map(lambda face: pontos_por_numero(dados, face), faces_presentes),
        (0,0)
    )

print(pontuacao_jogada([1, 5, 2, 3, 4, 6]))  # (1500, 6)
print(pontuacao_jogada([2, 2, 3, 4, 4, 6]))  # (0, 0)
print(pontuacao_jogada([1, 1, 1, 2, 3, 4]))  # (1000, 3)
print(pontuacao_jogada([2, 2, 2, 4, 5, 6]))  # (250, 4)
print(pontuacao_jogada([1, 5, 2, 2, 3, 4]))  # (150, 2)
print(pontuacao_jogada([2, 2, 2, 2, 2, 2]))  # (400, 6)
print(pontuacao_jogada([2, 2, 2, 3, 3, 3]))  # (500, 6)
print(pontuacao_jogada([4, 4, 4, 6, 6, 6]))  # (1000, 6)
print(pontuacao_jogada([5, 6, 5, 6, 6, 6]))  # (700, 5)
````

#### 2.6. Turno
Crie a função _recursiva_ ```turno(pontos_rodada, n_dados, escolha, verbose = True)``` que implementa um turno de jogadas de um jogador. Use as funções ```rolar_dados``` para rolar os dados; ```pontuacao_jogada``` para obter pontos da jogada e número de dados pontuadores; e a função ```escolha``` para definir a escolha do usuário, como descrito melhor abaixo. A função tem como parâmetros:
* ```pontos_rodada```: número de pontos acumulados pelo jogador até a rodada atual
* ```n_dados```: número de dados a serem jogados
* ```escolha```: função usada para decidir se o jogador para ou continua jogando na rodada. A decisão é por parar se ```escolha``` retorna ```False```. É continuar se ```escolha``` retorna ```True```. Esta função deve tomar a decisão com base no número de dados restantes e na pontuação atual do jogador. Abaixo, um exemplo de escolha aleatória (que, por ser aleatória, ignora os parâmetros :)):
```python
def escolha_aleatoria(n_restante, pontos):
    return random.choice([True, False])
```
* ```verbose = False```: parâmetro opcional que indica se mensagens de status serão exibidas ou não, indicando o que ocorre no turno.

````python
import random
def escolha_aleatoria(n_restante, pontos):
    return random.choice([True, False])
    

def turno(pontos_rodada, n_dados, escolha, verbose = True):
    rolamento_dados = rolar_dados(n_dados)
    pontos_referentes, qtd_dados = pontuacao_jogada(rolamento_dados)
    
    if pontos_referentes == 0:
        return 0
    
    pontos_acumulados = pontos_rodada + pontos_referentes
    def calcula_dados(total, usados):
        resto = total - usados
        return 6 if resto == 0 else resto
    
    dados_restantes = calcula_dados(n_dados, qtd_dados)
    
    if escolha(dados_restantes, pontos_acumulados) == True:
        return turno(pontos_acumulados, dados_restantes, escolha, verbose)
    else:
        return pontos_acumulados
````

#### 2.7. Escolha heurística
Crie a função ```escolha_heuristica(n_restante, pontos)``` que, ao contrário da função ```escolha_aleatória```, efetivamente usa a informação de número restante de dados e pontos atuais do jogador para escolhar se para ou continua. Projete uma heurística que tenha uma probabilidade maior de fazer mais pontos que a função de escolha aleatória ao longo do tempo.

````python
def escolha_heuristica(n_restante, pontos):
    if n_restante >= 4:
        return True
    elif n_restante == 3:
        return pontos < 300
    elif n_restante == 2:
        return pontos < 200
    elif n_restante == 1:
        return pontos < 150
    else:
        return False
````

```python
def partida(estrategia0, estrategia1, pontos=[0, 0], jogador=0, pontuacao_max=10000, verbose=False):
    '''
        * estrategia0: função de escolha para jogador 0
        * estrategia1: função de escolha para jogador 1
        * pontos = [0, 0]: pontuação do jogador 0 e do jogador 1 
          (0 e 0 no início da partida)
        * jogador = 0: jogador que vai iniciar a partida (0, por default)
        * pontuacao_max = 10000: pontuação máxima que determina fim de jogo 
          (10000 por default)
    '''
    # Condição de término
    if pontos[0] >= pontuacao_max or pontos[1] >= pontuacao_max:
        if verbose:
            print(f"\nFIM DE JOGO! Placar final: Jogador 0: {pontos[0]} | Jogador 1: {pontos[1]}")
        if pontos[0] >= pontuacao_max and pontos[1] >= pontuacao_max:
            if pontos[0] > pontos[1]:
                return 0
            elif pontos[1] > pontos[0]:
                return 1
            else:
                return None
        elif pontos[0] >= pontuacao_max:
            return 0
        else:
            return 1

    # Turno do jogador atual
    if verbose:
        print(f"\n- Jogador {jogador} (Placar: {pontos[0]}-{pontos[1]}) ---")

                    pontos[1] + (pts_rodada if jogador == 1 else 0)]
    
    return partida(estrategia0, estrategia1, novos_pontos, 
                   1 - jogador, pontuacao_max, verbose)
```
